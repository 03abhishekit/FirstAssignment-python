```python
Q1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Line Plot (sns.lineplot):

Use: Useful for visualizing data points in a continuous time series or relationship between two continuous variables.
Example: Visualizing stock prices over time.

Bar Plot (sns.barplot):

Use: Used to compare different categories or to show changes in values over time.
Example: Comparing the average scores of students in different classes.

Box Plot (sns.boxplot):

Use: Displays the distribution of data based on a five-number summary (minimum, first quartile, median, third quartile, and maximum).
Example: Comparing the distribution of ages among different groups.

Histogram (sns.histplot):

Use: Useful for displaying the distribution of a single variable, showing the frequency of observations within certain ranges.
Example: Showing the distribution of test scores in a class.

Heatmap (sns.heatmap):

Use: Used to show the magnitude of a phenomenon as color in two dimensions, useful for visualizing correlation matrices or showing frequency of occurrence.
Example: Displaying the correlation between different features in a dataset.






Q4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.


# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Create a histogram for 'price' with hue based on 'cut'
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack')
plt.title('Price Distribution by Cut')
plt.show()

Q5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.


# Load the iris dataset
iris = sns.load_dataset("iris")

# Create a pair plot with hue based on 'species'
sns.pairplot(iris, hue='species')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()

Q6: Use the "flights" dataset from seaborn to plot a heatmap.


# Load the flights dataset
flights = sns.load_dataset("flights")

# Pivot the dataset to get a matrix format
flights_pivot = flights.pivot("month", "year", "passengers")

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Heatmap of Flights Dataset')
plt.show()

Example Jupyter Notebook
Below is a summary of the code that should be included in your Jupyter notebook. Ensure that each code cell is properly annotated and that the plots are correctly displayed.



# Importing necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Q1: Explanation of different plots with examples

# Q2: Line Plot using fmri dataset
fmri = sns.load_dataset("fmri")
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title('FMRI Signal over Time')
plt.show()

# Q3: Box Plots using titanic dataset
titanic = sns.load_dataset("titanic")
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title('Age Distribution by Pclass')
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title('Fare Distribution by Pclass')
plt.tight_layout()
plt.show()

# Q4: Histogram using diamonds dataset
diamonds = sns.load_dataset("diamonds")
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack')
plt.title('Price Distribution by Cut')
plt.show()

# Q5: Pair Plot using iris dataset
iris = sns.load_dataset("iris")
sns.pairplot(iris, hue='species')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()

# Q6: Heatmap using flights dataset
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot("month", "year", "passengers")
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Heatmap of Flights Dataset')
plt.show()
```

```python
Q2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.


import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Create a line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title('FMRI Signal over Time')
plt.show()
```

```python
Q3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.


# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Create box plots for 'age' and 'fare' based on 'pclass'
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title('Age Distribution by Pclass')

plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title('Fare Distribution by Pclass')

plt.tight_layout()
plt.show()
```
