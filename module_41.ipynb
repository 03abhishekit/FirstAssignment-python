```python
Q1: Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.


import seaborn as sns
import plotly.express as px

# Load the titanic dataset using seaborn
titanic = sns.load_dataset("titanic")

# Plot a scatter plot using Plotly express
fig = px.scatter(titanic, x="age", y="fare", title="Age vs Fare Scatter Plot")
fig.show()

Q2: Using the tips dataset in the Plotly library, plot a box plot using Plotly express.


import plotly.express as px

# Load the tips dataset using Plotly express
tips = px.data.tips()

# Plot a box plot using Plotly express
fig = px.box(tips, x="day", y="total_bill", title="Box Plot of Total Bill by Day")
fig.show()

Q3: Using the tips dataset in the Plotly library, Plot a histogram for x="sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.


import plotly.express as px

# Load the tips dataset using Plotly express
tips = px.data.tips()

# Plot a histogram using Plotly express
fig = px.histogram(tips, x="sex", y="total_bill", color="day", pattern_shape="smoker", 
                   title="Histogram of Total Bill by Sex and Smoker Status",
                   barmode="group")
fig.show()

Q4: Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.


import plotly.express as px

# Load the iris dataset using Plotly express
iris = px.data.iris()

# Plot a scatter matrix plot using Plotly express
fig = px.scatter_matrix(iris,
                        dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
                        color="species",
                        title="Scatter Matrix Plot of Iris Dataset")
fig.show()

Q5: What is Distplot? Using Plotly express, plot a distplot.
Distplot: A distplot (distribution plot) is used to visualize the distribution of a dataset. It is a combination of a histogram and a kernel density estimate (KDE).

Using Plotly express, you can use px.histogram with the marginal parameter set to 'rug' or 'violin' to create a distplot-like visualization.

python
Copy code
import plotly.express as px

# Load the tips dataset using Plotly express
tips = px.data.tips()

# Plot a distplot-like histogram using Plotly express
fig = px.histogram(tips, x="total_bill", marginal="rug", title="Distplot of Total Bill")
fig.show()
am(tips, x="total_bill", marginal="rug", title="Distplot of Total Bill")
fig5.show()
```
