Q1: What is the role of feature selection in anomaly detection?
Role:

Dimensionality Reduction: Reduces the number of features to improve the efficiency and effectiveness of anomaly detection algorithms.
Noise Reduction: Eliminates irrelevant or redundant features that may introduce noise and hinder anomaly detection.
Improved Performance: Enhances the performance of anomaly detection algorithms by focusing on relevant features that capture the underlying patterns of normal and anomalous behavior.


Q2: What are some common evaluation metrics for anomaly detection algorithms and how are they computed?
Common Evaluation Metrics:

Accuracy: Proportion of correctly classified instances (for supervised anomaly detection).
Precision: Ratio of correctly identified anomalies to the total number of anomalies detected.
Recall: Ratio of correctly identified anomalies to the total number of true anomalies.
F1 Score: Harmonic mean of precision and recall.
Area Under the ROC Curve (AUC-ROC): Measures the model's ability to distinguish between normal and anomalous instances across different thresholds.
Precision-Recall Curve (PRC): Graphical representation of precision and recall for varying threshold values.
Confusion Matrix: Tabulates true positives, true negatives, false positives, and false negatives to evaluate classification performance.

Q8: What are local outliers and global outliers, and how do they differ from each other?
Local Outliers: Data points that are considered outliers within a local neighborhood but may not be outliers when considering the entire dataset.
Global Outliers: Data points that are considered outliers across the entire dataset.
Difference:

Scope: Local outliers are outliers within a specific context or neighborhood, while global outliers are outliers across the entire dataset.
Detection: Local outliers are detected based on local characteristics, while global outliers are detected based on global characteristics.

Q10: How can global outliers be detected using the Isolation Forest algorithm?
Isolation Forest:

Concept: Randomly partitions the feature space and isolates outliers in fewer partitions, making them easier to detect.
Detection:
Randomly select features and split data points until each point is isolated in its own partition (tree).
Anomalies are identified as points with shorter average path lengths in the trees, indicating they are easier to isolate and hence more likely to be outliers.

Network Security: Identifying unusual patterns in local network traffic that may indicate intrusions or anomalies.
