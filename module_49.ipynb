```python
Q1: Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.

import scipy.stats as stats

# Given data
sample_mean = 50
sample_std_dev = 5
sample_size = 30  # Assuming a sample size

# Calculate the 95% confidence interval
confidence_level = 0.95
degrees_freedom = sample_size - 1
confidence_interval = stats.t.interval(confidence_level, degrees_freedom, sample_mean, sample_std_dev / (sample_size ** 0.5))

print(f"95% Confidence Interval: {confidence_interval}")
Interpretation:
The 95% confidence interval provides a range within which we can be 95% confident that the true population mean lies. 
If the interval is (48.20, 51.80), it means we are 95% confident that the population mean is between 48.20 and 51.80.


Q2: Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&amp;
Ms in a bag matches the expected distribution using Python.

import scipy.stats as stats
import numpy as np

# Observed frequencies
observed = np.array([30, 25, 25, 10, 10, 20])

# Expected frequencies
total = observed.sum()
expected = np.array([0.20, 0.20, 0.20, 0.10, 0.10, 0.20]) * total

# Chi-square test
chi2_stat, p_val = stats.chisquare(observed, expected)

print(f"Chi-square Statistic: {chi2_stat}, p-value: {p_val}")

# Interpretation
alpha = 0.05
if p_val &lt; alpha:
    print("Reject the null hypothesis: The observed distribution does not match the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed distribution matches the expected distribution.")
    
    
Q3: Use Python to calculate the chi-square statistic and p-value for a contingency table.

# Given contingency table
contingency_table = np.array([[20, 15], [10, 25], [15, 20]])

# Chi-square test
chi2_stat, p_val, dof, expected = stats.chi2_contingency(contingency_table)

print(f"Chi-square Statistic: {chi2_stat}, p-value: {p_val}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)

# Interpretation
alpha = 0.05
if p_val &lt; alpha:
    print("Reject the null hypothesis: There is a significant association between the variables.")
else:
    print("Fail to reject the null hypothesis: There is no significant association between the variables.")
    
    
Q4: Calculate the 95% confidence interval for the true proportion of individuals in the population who smoke using Python.

# Given data
sample_proportion = 60 / 500
sample_size = 500

# Calculate the standard error
standard_error = np.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)

# Calculate the 95% confidence interval
confidence_level = 0.95
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)
margin_of_error = z_score * standard_error

lower_bound = sample_proportion - margin_of_error
upper_bound = sample_proportion + margin_of_error

print(f"95% Confidence Interval: ({lower_bound}, {upper_bound})")
Interpretation:
The confidence interval provides a range within which the true proportion of smokers in the population is expected to lie with 95% confidence.

Q5: Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12 using Python.

# Given data
sample_mean = 75
sample_std_dev = 12
sample_size = 30  # Assuming a sample size

# Calculate the 90% confidence interval
confidence_level = 0.90
degrees_freedom = sample_size - 1
confidence_interval = stats.t.interval(confidence_level, degrees_freedom, sample_mean, sample_std_dev / (sample_size ** 0.5))

print(f"90% Confidence Interval: {confidence_interval}")
Interpretation:
The 90% confidence interval provides a range within which we can be 90% confident that the true population mean lies.

Q6: Use Python to plot the chi-square distribution with 10 degrees of freedom.
Label the axes and shade the area corresponding to a chi-square statistic of 15.

import matplotlib.pyplot as plt
import numpy as np

# Degrees of freedom
df = 10

# Generate x values
x = np.linspace(0, 30, 1000)

# Chi-square distribution
y = stats.chi2.pdf(x, df)

# Plot
plt.plot(x, y, label=f'Chi-square distribution (df={df})')
plt.fill_between(x, 0, y, where=(x &lt;= 15), color='skyblue', alpha=0.5)
plt.axvline(x=15, color='red', linestyle='--', label='Chi-square statistic = 15')
plt.title('Chi-square Distribution with 10 Degrees of Freedom')
plt.xlabel('Chi-square value')
plt.ylabel('Probability density')
plt.legend()
plt.show()


Q7: Calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke using Python.

# Given data
sample_proportion = 520 / 1000
sample_size = 1000

# Calculate the standard error
standard_error = np.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)

# Calculate the 99% confidence interval
confidence_level = 0.99
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)
margin_of_error = z_score * standard_error

lower_bound = sample_proportion - margin_of_error
upper_bound = sample_proportion + margin_of_error

print(f"99% Confidence Interval: ({lower_bound}, {upper_bound})")
Q8: Conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin using Python.pythonCopy code# Given data
observed = np.array([45, 55])
expected = np.array([50, 50])  # Expected frequencies for a fair coin

# Chi-square test
chi2_stat, p_val = stats.chisquare(observed, expected)

print(f"Chi-square Statistic: {chi2_stat}, p-value: {p_val}")

# Interpretation
alpha = 0.05
if p_val &lt; alpha:
    print("Reject the null hypothesis: The coin is biased.")
else:
    print("Fail to reject the null hypothesis: The coin is fair.")
Q9: Conduct a chi-square test for independence to determine if there is a significant association between smoking status and lung cancer diagnosis using Python.pythonCopy code# Given contingency table
contingency_table = np.array([[60, 140], [30, 170]])

# Chi-square test
chi2_stat, p_val, dof, expected = stats.chi2_contingency(contingency_table)

print(f"Chi-square Statistic: {chi2_stat}, p-value: {p_val}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)

# Interpretation
alpha = 0.05
if p_val &lt; alpha:
    print("Reject the null hypothesis: There is a significant association between smoking status and lung cancer diagnosis.")
else:
    print("Fail to reject the null hypothesis: There is no significant association between smoking status and lung cancer diagnosis.")
Q10: Conduct a chi-square test for independence to determine if there is a significant association between chocolate preference and country of origin using Python.pythonCopy code# Given contingency table
contingency_table = np.array([[200, 150, 150], [180, 160, 160]])

# Chi-square test
chi2_stat, p_val, dof, expected = stats.chi2_contingency(contingency_table)

print(f"Chi-square Statistic: {chi2_stat}, p-value: {p_val}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)

# Interpretation
alpha = 0.01
if p_val &lt; alpha:
    print("Reject the null hypothesis: There is a significant association between chocolate preference and country of origin.")
else:
    print("Fail to reject the null hypothesis: There is no significant association between chocolate preference and country of origin.")
Q11: Conduct a hypothesis test to determine if the population mean is significantly different from 70 using Python.pythonCopy code# Given data
sample_mean = 72
sample_std_dev = 10
sample_size = 30
population_mean = 70

# Calculate the t-statistic
t_statistic = (sample_mean - population_mean) / (sample_std_dev / np.sqrt(sample_size))

# Calculate the p-value
p_val = stats.t.sf(np.abs(t_statistic), df=sample_size-1) * 2

print(f"T-statistic: {t_statistic}, p-value: {p_val}")

# Interpretation
alpha = 0.05
if p_val &lt; alpha:
    print("Reject the null hypothesis: The population mean is significantly different from 70.")
else:
    print("Fail to reject the null hypothesis: The population mean is not significant
```
