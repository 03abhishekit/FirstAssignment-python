```python
Q1. Load the flight price dataset and examine its dimensions. How many rows and columns does the dataset have?

import pandas as pd

# Load the dataset
flight_price_df = pd.read_csv('flight_price_dataset.csv')

# Display the dimensions of the dataset
rows, columns = flight_price_df.shape
print("Number of rows:", rows)
print("Number of columns:", columns)

Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the distribution.

import matplotlib.pyplot as plt

# Plot histogram of flight prices
plt.figure(figsize=(10, 6))
plt.hist(flight_price_df['Price'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Flight Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()


Q3. What is the range of prices in the dataset? What is the minimum and maximum price?

# Calculate the range of prices
price_range = flight_price_df['Price'].max() - flight_price_df['Price'].min()
min_price = flight_price_df['Price'].min()
max_price = flight_price_df['Price'].max()

print("Range of prices:", price_range)
print("Minimum price:", min_price)
print("Maximum price:", max_price)


Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different airlines.

# Create a boxplot to compare prices by airline
plt.figure(figsize=(12, 8))
sns.boxplot(x='Airline', y='Price', data=flight_price_df)
plt.title('Flight Prices by Airline')
plt.xlabel('Airline')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()


Q5. Are there any outliers in the dataset? Identify any potential outliers using a boxplot and describe how they may impact your analysis.

# Identify potential outliers using boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x=flight_price_df['Price'])
plt.title('Boxplot of Flight Prices')
plt.xlabel('Price')
plt.show()


Q6. You are working for a travel agency, and your boss has asked you to analyze the Flight Price dataset to identify the peak travel season.
What features would you analyze to identify the peak season, and how would you present your findings to your boss?

To identify the peak travel season, I would analyze the 'Date_of_Journey' feature to understand the trends in flight prices over time. 
I would aggregate the data by month or quarter and calculate the average flight prices for each period.
Then, I would visualize the average prices over time using line plots or bar plots.
Peaks in average prices would indicate the peak travel seasons.


Q7. You are a data analyst for a flight booking website, and you have been asked to analyze the Flight Price dataset to identify 
any trends in flight prices. What features would you analyze to identify these trends, and what visualizations would you use to present
your findings to your team?


I would analyze the 'Date_of_Journey', 'Source', 'Destination', 'Airline', and 'Total_Stops' features to identify trends in flight prices.
I would use line plots, bar plots, and heatmaps to visualize the average prices based on different combinations of these features over time.


Q8. You are a data scientist working for an airline company, and you have been asked to analyze the Flight Price dataset to identify
the factors that affect flight prices. What features would you analyze to identify these factors, and how would you present your findings
to the management team?

I would analyze the 'Airline', 'Source', 'Destination', 'Total_Stops', 'Additional_Info', and 'Duration' features to identify the factors
affecting flight prices. I would use statistical tests such as ANOVA or regression analysis to quantify the relationship between these 
features and flight prices. Additionally, I would create visualizations such as box plots, scatter plots, and heatmaps to illustrate these
relationships. Finally, I would provide a comprehensive report outlining the key factors influencing flight prices and their impact on 
pricing strategies.


Q9. Load the Google Playstore dataset and examine its dimensions. How many rows and columns does the dataset have?

import pandas as pd

# Load the dataset
playstore_df = pd.read_csv('google_playstore_dataset.csv')

# Display the dimensions of the dataset
rows, columns = playstore_df.shape
print("Number of rows:", rows)
print("Number of columns:", columns)


Q10. How does the rating of apps vary by category? Create a boxplot to compare the ratings of different app categories.

import matplotlib.pyplot as plt
import seaborn as sns

# Create a boxplot to compare ratings by category
plt.figure(figsize=(12, 8))
sns.boxplot(x='Category', y='Rating', data=playstore_df)
plt.title('App Ratings by Category')
plt.xlabel('Category')
plt.ylabel('Rating')
plt.xticks(rotation=90)
plt.show()


Q11. Are there any missing values in the dataset? Identify any missing values and describe how they may impact your analysis.

# Check for missing values
missing_values = playstore_df.isnull().sum()

# Display missing values
print("Missing Values:\n", missing_values)

# Describe the impact of missing values
impact_description = """ Missing values can impact analysis by introducing bias or reducing the accuracy of statistical models. 
In this dataset, missing values may affect insights related to app ratings, prices, or other features, 
especially if they are not handled properly through imputation or removal.
"""
print("Impact of Missing Values:\n", impact_description)


Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.

# Create a scatter plot to visualize the relationship between app size and rating
plt.figure(figsize=(10, 6))
plt.scatter(playstore_df['Size'], playstore_df['Rating'], alpha=0.5)
plt.title('Relationship between App Size and Rating')
plt.xlabel('Size')
plt.ylabel('Rating')
plt.show()


Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

# Create a bar chart to compare average prices by app type
plt.figure(figsize=(10, 6))
avg_price_by_type = playstore_df.groupby('Type')['Price'].mean().reset_index()
sns.barplot(x='Type', y='Price', data=avg_price_by_type)
plt.title('Average Price by App Type')
plt.xlabel('Type')
plt.ylabel('Average Price')
plt.show()


Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

# Create a frequency table to identify the top 10 most popular apps by installs

top_10_apps = playstore_df.groupby('App')['Installs'].sum().nlargest(10)
print("Top 10 Most Popular Apps by Installs:\n", top_10_apps)


Q15. A company wants to launch a new app on the Google Playstore and has asked you to analyze the Google Playstore dataset to identify 
the most popular app categories. How would you approach this task, and what features would you analyze to make recommendations to the company?To identify the most popular app categories, I would analyze the 'Category' feature and calculate the total number of installs for each category. Then, I would rank the categories based on the total number of installs and recommend the top categories with the highest popularity to the company. Additionally, I may also consider analyzing other features such as 'Rating', 'Reviews', and 'Size' to provide more insights into user preferences and trends within each category.3.5



```
