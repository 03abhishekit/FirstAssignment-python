```python
# Q1: What is Statistics?


Statistics is a branch of mathematics dealing with the collection, analysis, interpretation, presentation, and organization of data. 
It provides tools and techniques to make decisions based on data and to infer properties of the entire population from a sample.


# Q2: Define the different types of statistics and give an example of when each type might be used.

Types of Statistics:
1. Descriptive Statistics:
   - Definition: Descriptive statistics summarize and describe the features of a dataset.
   - Example: Calculating the average test score of a class.

2. Inferential Statistics:
   - Definition: Inferential statistics make predictions or inferences about a population based on a sample of data drawn from the population.
   - Example: Using a sample of voter preferences to predict the outcome of an election.


# Q3: What are the different types of data and how do they differ from each other? Provide an example of each type of data.

Types of Data:
1. Qualitative (Categorical) Data:
   - Definition: Data that can be categorized based on traits and labels.
   - Example: Colors of a set of balls (red, blue, green).

2. Quantitative Data:
   - Definition: Data that can be measured and expressed numerically.
   - Example: Heights of students in a class (150 cm, 160 cm, 170 cm).


# Q5: Explain the concept of levels of measurement and give an example of a variable for each level.

Levels of Measurement:
1. Nominal:
   - Definition: Categories without a specific order.
   - Example: Gender (Male, Female).

2. Ordinal:
   - Definition: Categories with a specific order.
   - Example: Education level (High School, Bachelor's, Master's, PhD).

3. Interval:
   - Definition: Numeric data without a true zero point.
   - Example: Temperature in Celsius.

4. Ratio:
   - Definition: Numeric data with a true zero point.
   - Example: Weight (0 kg means no weight).


# Q6: Why is it important to understand the level of measurement when analyzing data? Provide an example to illustrate your answer.

Understanding the level of measurement is important because it determines the types of statistical analyses that can be performed. For example, calculating the mean makes sense for interval and ratio data but not for nominal data. 

Example: You can calculate the average income (ratio data) of a group, but calculating an average gender (nominal data) doesn't make sense.


# Q7: How nominal data type is different from ordinal data type.

Nominal data represent categories without any intrinsic order, while ordinal data represent categories with a specific order.

Example:
- Nominal: Types of fruit (apple, orange, banana).
- Ordinal: Movie ratings (poor, fair, good, excellent).


# Q8: Which type of plot can be used to display data in terms of range?

A box plot (box-and-whisker plot) can be used to display data in terms of range, showing the minimum, first quartile, median, third quartile, and maximum.


# Q9: Describe the difference between descriptive and inferential statistics. Give an example of each type of statistics and explain how they are used.

Descriptive Statistics:
- Definition: Summarize and describe the features of a dataset.
- Example: Mean, median, and standard deviation of test scores.
- Usage: Provides a summary of the data.

Inferential Statistics:
- Definition: Make predictions or inferences about a population based on a sample.
- Example: Hypothesis testing, confidence intervals.
- Usage: Allows for generalization from a sample to a population.
"""

# Q10: What are some common measures of central tendency and variability used in statistics? Explain how each measure can be used to describe a dataset.
"""
Measures of Central Tendency:
1. Mean: The average of a dataset.
2. Median: The middle value of a dataset when sorted.
3. Mode: The most frequently occurring value in a dataset.

Measures of Variability:
1. Range: The difference between the maximum and minimum values.
2. Variance: The average of the squared differences from the mean.
3. Standard Deviation: The square root of the variance, representing the average distance from the mean.

These measures provide insights into the typical values and the spread of the data.
"""

# Q4: Categorise the following datasets with respect to quantitative and qualitative data types:
grading = ["A+", "A", "B+", "B", "C+", "C", "D", "E"]
colors = ["yellow", "green", "orange", "red"]
heights = [178.9, 179, 179.5, 176, 177.2, 178.3, 175.8]
number_of_mangoes = [500, 600, 478, 672]

grading_type = "Qualitative"
colors_type = "Qualitative"
heights_type = "Quantitative"
number_of_mangoes_type = "Quantitative"

categorization = {
    "Grading in exam": grading_type,
    "Colour of mangoes": colors_type,
    "Height data of a class": heights_type,
    "Number of mangoes exported by a farm": number_of_mangoes_type
}

categorization
```
