```python
Q1: Pearson Correlation Coefficient
Task: Calculate the Pearson correlation coefficient between the amount of time students spend studying for an exam and
their final exam scores.
Example Data:

import numpy as np
import pandas as pd
from scipy.stats import pearsonr

# Example Data
study_hours = [10, 9, 3, 5, 7, 4, 6, 8, 2, 1]
exam_scores = [95, 90, 70, 80, 85, 60, 75, 92, 50, 45]

# Calculate Pearson Correlation Coefficient
pearson_corr, _ = pearsonr(study_hours, exam_scores)
print(f'Pearson Correlation Coefficient: {pearson_corr}')
Interpretation: A Pearson correlation coefficient close to 1 indicates a strong positive linear relationship, 
close to -1 indicates a strong negative linear relationship, and close to 0 indicates no linear relationship.

Q2: Spearman's Rank CorrelationTask: 
Calculate the Spearman's rank correlation between the amount of sleep individuals get each night and their overall job satisfaction level.

Example Data:
    
from scipy.stats import spearmanr

# Example Data
sleep_hours = [7, 8, 5, 6, 7, 8, 9, 4, 6, 5]
job_satisfaction = [8, 9, 6, 7, 8, 10, 9, 5, 6, 7]

# Calculate Spearman's Rank Correlation
spearman_corr, _ = spearmanr(sleep_hours, job_satisfaction)
print(f'Spearman\'s Rank Correlation Coefficient: {spearman_corr}')
Interpretation: A Spearman correlation coefficient close to 1 or -1 indicates a strong monotonic relationship,
while a value close to 0 indicates no monotonic relationship.

Q3: Pearson vs. Spearman Correlation
Task: Calculate both the Pearson correlation coefficient and Spearman's rank correlation between the number of hours of exercise per week
and BMI for 50 participants and compare the results.

Example Data:
# Generate random data
np.random.seed(0)
exercise_hours = np.random.randint(0, 10, 50)
bmi = np.random.normal(25, 5, 50)

# Calculate Pearson Correlation Coefficient
pearson_corr, _ = pearsonr(exercise_hours, bmi)
print(f'Pearson Correlation Coefficient: {pearson_corr}')

# Calculate Spearman's Rank Correlation
spearman_corr, _ = spearmanr(exercise_hours, bmi)
print(f'Spearman\'s Rank Correlation Coefficient: {spearman_corr}')
Comparison: Compare the values of the two coefficients to understand the nature of the relationship (linear vs. monotonic).

Q4: Pearson Correlation for TV Hours and Physical Activity
Task: Calculate the Pearson correlation coefficient between the number of hours individuals spend watching television
per day and their level of physical activity.

Example Data:

# Generate random data
np.random.seed(0)
tv_hours = np.random.randint(1, 5, 50)
physical_activity = np.random.randint(1, 10, 50)

# Calculate Pearson Correlation Coefficient
pearson_corr, _ = pearsonr(tv_hours, physical_activity)
print(f'Pearson Correlation Coefficient: {pearson_corr}')


Q5: Survey on Age and Soft Drink Preference

Task: Survey results data not provided.
Typically, one would encode categorical data and calculate correlation.

Step 1: Prepare the DataFirst, lets put the data into a pandas DataFrame.
Step 2: Perform Label EncodingHeres the complete code to perform label encoding using the scikit-learn library:

import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Creating the DataFrame
data = {
    'Age(Years)': [25, 42, 37, 19, 31, 28],
    'Preference': ['Coke', 'Pepsi', 'Mountain Dew', 'Coke', 'Pepsi', 'Coke']
}

df = pd.DataFrame(data)

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the Preference column
df['Preference_Encoded'] = label_encoder.fit_transform(df['Preference'])

# Display the DataFrame with the encoded column
print(df)
print("\nLabel Encoding Mapping:")
for class_index in range(len(label_encoder.classes_)):
    print(f'{label_encoder.classes_[class_index]}: {class_index}')
    
ExplanationData Preparation:We created a DataFrame using the given data.Label Encoding:We initialized a LabelEncoder from scikit-learn.We fit the label encoder on the Preference column and transformed it into numerical labels.We printed out the DataFrame with the new encoded column and the mapping of the original labels to the encoded labels.

Q6: Sales Calls and Sales MadeTask: Calculate the Pearson correlation coefficient between the number of sales calls made per day and the number of sales made per week.Example Data:pythonCopy code# Generate random data
np.random.seed(0)
sales_calls = np.random.randint(10, 50, 30)
sales_made = np.random.randint(1, 15, 30)

# Calculate Pearson Correlation Coefficient
pearson_corr, _ = pearsonr(sales_calls, sales_made)
print(f'Pearson Correlation Coefficient: {pearson_corr}')
```
